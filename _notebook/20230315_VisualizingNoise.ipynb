Visualizing Random Noise
===

Visualizing random noise.

Intended to create a silly little thumbnail.

```python
# numpy and pandas for data-wrangling
import numpy as np
import pandas as pd
# matplotlib for plotting
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 300
```

```python
rng = np.random.default_rng()
rng.normal(loc=5, scale=1, size=10)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))

n = 20000
xs = rng.normal(loc=50, scale=1.6, size=n)
ys = rng.normal(loc=50, scale=1.3, size=n) * xs * 2
ax.hexbin(xs, ys, bins='log', xscale='log', yscale='log', mincnt=1)

fig.tight_layout()
plt.show()
```
